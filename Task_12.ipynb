```python
import pandas as pd
import numpy as np
# Create a DataFrame from the provided data
data = {
    'education': ['Bachelors', 'Bachelors', 'HS-grad', '11th', 'Bachelors'],
    'age': [39, 50, 38, 53, 28],
    'capital-gain': [2174, '?', '?', '?', 0],
    'race': ['White', 'White', 'White', 'Black', 'Black'],
    'capital-loss': [0, 0, 0, 0, 0],
    'hours-per-week': [40, 13, 40, 40, 40],
    'gender': [1, 1, 1, 1, 0],
    'classification': ['<=50K', '<=50K', '<=50K', '<=50K', '<=50K']
}

df = pd.DataFrame(data)

# Question 1: How many rows and columns are there in the dataset?
num_rows, num_columns = df.shape
print(f"Question 1: There are {num_rows} rows and {num_columns} columns in the dataset.")

# Question 2: How many missing values are there in the 'capital-gain' column?
missing_values = df['capital-gain'].str.contains('\?').sum()
print(f"Question 2: There are {missing_values} missing values in the 'capital-gain' column.")

# Question 3: What is the average age of individuals in the dataset?
average_age = df['age'].mean()
print(f"Question 3: The average age of individuals in the dataset is {average_age} years.")

# Question 4: How many unique education levels are there in the dataset?
unique_education_levels = df['education'].nunique()
print(f"Question 4: There are {unique_education_levels} unique education levels in the dataset.")

# Question 5: What is the most common race in the dataset?
most_common_race = df['race'].mode()[0]
print(f"Question 5: The most common race in the dataset is '{most_common_race}'.")

# Question 6: How many records belong to the 'Male' gender category?
male_count = df[df['gender'] == 1].shape[0]
print(f"Question 6: There are {male_count} records with 'Male' gender category.")

# Question 7: What is the maximum value in the 'hours-per-week' column?
max_hours_per_week = df['hours-per-week'].max()
print(f"Question 7: The maximum value in the 'hours-per-week' column is {max_hours_per_week}.")

# Question 8: How many unique values are there in the 'classification' column?
unique_classification_values = df['classification'].nunique()
print(f"Question 8: There are {unique_classification_values} unique values in the 'classification' column.")

# Question 9: How many individuals have 'capital-loss' greater than 0?
capital_loss_greater_than_0 = df[df['capital-loss'] > 0].shape[0]
print(f"Question 9: {capital_loss_greater_than_0} individuals have 'capital-loss' greater than 0.")



# Question 10: What is the average age for each unique education level?
average_age_by_education = df.groupby('education')['age'].mean()
print("Question 10: Average age by education level:")
print(average_age_by_education)

# Question 11: How many individuals belong to each 'classification' category within each 'race'?
classification_counts_by_race = df.groupby(['race', 'classification']).size().unstack()
print("Question 11: Classification counts by race:")
print(classification_counts_by_race)

# Question 12: What is the maximum 'hours-per-week' worked by individuals in each 'gender' category?
max_hours_per_week_by_gender = df.groupby('gender')['hours-per-week'].max()
print("Question 12: Maximum hours-per-week by gender:")
print(max_hours_per_week_by_gender)

# Question 13: What is the average 'age' and 'hours-per-week' worked by each 'gender' within each 'classification' category?
average_age_hours_by_gender_classification = df.groupby(['gender', 'classification'])[['age', 'hours-per-week']].mean()
print("Question 13: Average age and hours-per-week by gender and classification:")
print(average_age_hours_by_gender_classification)





```
