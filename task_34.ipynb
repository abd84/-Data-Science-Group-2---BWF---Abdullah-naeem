```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import tensorflow as tf
from tensorflow import keras

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
print(X)
# print(y)


encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y.reshape(-1, 1))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**Input Layer:** The number of neurons in the input layer should match the number of features in your dataset.

**Output Layer:** The number of neurons should match the number of output classes or the number of output values you are predicting.


Basic Regression Problem: For predicting a single continuous variable, you might start with a small network, like **[input -> 8 neurons -> 4 neurons -> output].**
Binary Classification: For classifying data into two categories, a network with **[input -> 16 neurons -> 8 neurons -> output]** is often a good starting point.
Image Recognition: For more complex tasks like image recognition, you might use more neurons per layer, like **[input -> 128 neurons -> 64 neurons -> 32 neurons -> output].**

# Epochs
Epochs are simply the number of times the entire dataset passes forward and backpropagated by updating the weights on a neural network. BY doing this we can find unseen patterns and information in every single epoch and hence it improves the accuracy of the model.

# Losses
Loss functions are a very important thing to notice while creating a neural network because loss functions in the neural network will calculate the difference between the predicted output and the actual result

```python

model = keras.Sequential([
    keras.layers.Dense(16, input_shape=(4,), activation='relu'),  # Hidden layer
    keras.layers.Dense(8, activation='relu'),                     # Additional hidden layer
    keras.layers.Dense(3, activation='softmax')                   # Output layer (3 classes for Iris species)
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=50, validation_split=0.2, verbose=1)


loss, accuracy = model.evaluate(X_test, y_test, verbose=1)
print(f'Test accuracy: {accuracy:.2f}')
predictions = model.predict(X_test)
print(predictions)
```
