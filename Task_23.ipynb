```python
import numpy as np

np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)


w = np.random.randn(1)
b = np.random.randn(1)
learning_rate = 0.01
n_iterations = 1000
m = len(X)
```

# 2. Hypothesis Function

```python
def predict(X, w, b):
    return X.dot(w) + b
```

# Mean Squared Error (MSE)
The loss function commonly used for linear regression is the Mean Squared



```python
def compute_mse(y_pred, y):
    return np.mean((y_pred - y) ** 2)
```

# Gradient decesent
To minimize the MSE, we use gradient descent. The gradients of the loss function with respect to

w and b are computed and used to update the parameters.

The update rules for gradient descent are:

Where

 is the learning rate.



```python
for iteration in range(n_iterations):
    y_pred = predict(X, w, b)
    error = y_pred - y

    # Calculate MSE
    mse = compute_mse(y_pred, y)

    # Gradients
    dw = (2/m) * X.T.dot(error)
    db = (2/m) * np.sum(error)

    # Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

    if iteration % 100 == 0:
        print(f"Iteration {iteration}: MSE = {mse}")
        print(f"Previous values: {y[:5].flatten()}")
        print(f"Predicted values: {y_pred[:5].flatten()}\n")

print(f"Final weights: {w}, Final bias: {b}")
```
