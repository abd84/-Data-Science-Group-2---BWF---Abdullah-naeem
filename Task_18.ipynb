```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv('Iris.csv')
df.head()
```

# Scatter Plot

A scatter plot displays values for two variables as points. Each point represents an observation in the dataset. Scatter plots are useful for visualizing the relationship between two continuous variables and can reveal patterns, correlations, and clusters

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

# Line Plot

A line plot displays data points connected by straight lines. It is used to visualize trends over time or a sequence of data points. Line plots are especially useful for time series data or any sequential data.

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10, 6))
sns.lineplot(x='Id', y='SepalLengthCm', hue='Species', data=df)
plt.title('Line Plot of Sepal Length Over Id')
plt.xlabel('Id')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

# Bar Plot

A bar plot displays the distribution of categorical data using rectangular bars. The height of each bar represents the count or value of the corresponding category. Bar plots are useful for comparing quantities among different groups or categories.



```python
sns.set_style('dark')
plt.figure(figsize=(10, 6))
sns.barplot(x='Species', y='SepalLengthCm', palette='plasma', data=df)
plt.title('Bar Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

# Count Plot

A count plot is a special type of bar plot that shows the count of observations in each categorical bin. It is useful for displaying the frequency distribution of a categorical variable.



```python
sns.set_style('white')
plt.figure(figsize=(10, 6))
sns.countplot(x='Species', data=df)
plt.title('Count Plot of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

# Box Plot

A box plot displays the distribution of a dataset based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It highlights the central tendency, variability, and outliers in the data. Box plots are useful for comparing distributions across different groups.

```python
sns.set_style('dark')
plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```
