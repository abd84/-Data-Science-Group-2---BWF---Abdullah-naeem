```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('titanic.csv')
df.head(10)
```

# Create Subplots (2x2)

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(df['PassengerId'], df['Fare'], marker='o')
axs[0, 0].set_title('Line Plot of Fare by PassengerId')
axs[0, 0].set_xlabel('PassengerId')
axs[0, 0].set_ylabel('Fare')

# Scatter plot
axs[0, 1].scatter(df['Age'], df['Fare'], c=df['Survived'], cmap='viridis')
axs[0, 1].set_title('Scatter Plot of Age vs Fare')
axs[0, 1].set_xlabel('Age')
axs[0, 1].set_ylabel('Fare')

# Bar plot
df['Survived'].value_counts().plot(kind='bar', ax=axs[1, 0], color=['blue', 'orange'])
axs[1, 0].set_title('Bar Plot of Survived Counts')
axs[1, 0].set_xlabel('Survived')
axs[1, 0].set_ylabel('Count')

# Histogram
axs[1, 1].hist(df['Age'].dropna(), bins=10, color='green', edgecolor='black')
axs[1, 1].set_title('Histogram of Age')
axs[1, 1].set_xlabel('Age')
axs[1, 1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Create Histograms

```python
# Single histogram with custom number of bins
plt.figure(figsize=(10, 6))
plt.hist(df['Fare'], bins=20, color='purple', edgecolor='black')
plt.title('Histogram of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

# Multiple histograms on the same figure
plt.figure(figsize=(10, 6))
plt.hist(df[df['Survived'] == 0]['Age'].dropna(), bins=10, alpha=0.5, label='Not Survived', color='red', edgecolor='black')
plt.hist(df[df['Survived'] == 1]['Age'].dropna(), bins=10, alpha=0.5, label='Survived', color='blue', edgecolor='black')
plt.title('Histogram of Age by Survival')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()


# Create normalized histograms
plt.figure(figsize=(10, 6))
plt.hist(df['Age'].dropna(), bins=10, density=True, alpha=0.5, label='Age', color='blue')
plt.hist(df['Fare'], bins=10, density=True, alpha=0.5, label='Fare', color='red')

# Add labels and titles
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normalized Histograms of Age and Fare')
plt.legend()
plt.show()

```

# Create Box Plots

```python
# Single box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Fare'])
plt.title('Box Plot of Fare')
plt.xlabel('Fare')
plt.show()

# Multiple box plots in a single figure
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=df, palette='Set2')
plt.title('Box Plot of Age by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

# Customized box plots
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=df, palette='Set3')
plt.title('Box Plot of Fare by Pclass and Survival')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```
