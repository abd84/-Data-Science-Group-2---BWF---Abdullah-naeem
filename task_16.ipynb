```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Housing.csv')
df.head()
```

## 1. Simple Line Plot

```python


plt.figure(figsize=(10, 6))
plt.plot(df['price'], label='Price')
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Simple Line Plot of Prices')
plt.legend()
plt.show()
```

## 2. Multiple Lines on the Same Graph

```python
plt.figure(figsize=(10, 6))
plt.plot(df['price'], label='Price')
plt.plot(df['area'], label='Area')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Line Plot of Prices and Areas')
plt.legend()
plt.show()
```

## 3. Line Plot with Markers

```python
plt.figure(figsize=(10, 6))
plt.plot(df['price'], marker='o', linestyle='-', color='b', label='Price')
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Line Plot with Markers')
plt.legend()
plt.show()
```

## 4. Customized Line Styles and Colors

```python
plt.figure(figsize=(10, 6))
plt.plot(df['price'], marker='x', linestyle='--', color='r', label='Price')
plt.plot(df['area'], marker='s', linestyle='-.', color='g', label='Area')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Customized Line Styles and Colors')
plt.legend()
plt.show()
```

## 5. Basic Scatter Plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['area'], df['price'], color='b', label='Price vs Area')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Scatter Plot of Price vs Area')
plt.legend()
plt.show()
```

## 6. Customized Scatter Plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['area'], df['price'], color='r', marker='x', s=100, label='Price vs Area')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Customized Scatter Plot of Price vs Area')
plt.legend()
plt.show()
```

## 7. Multiple Scatter Plots on the Same Graph

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['area'], df['price'], color='b', marker='o', s=50, label='Price vs Area')
plt.scatter(df['bedrooms'], df['price'], color='r', marker='x', s=100, label='Price vs Bedrooms')
plt.xlabel('Feature')
plt.ylabel('Price')
plt.title('Multiple Scatter Plots on the Same Graph')
plt.legend()
plt.show()
```

## 8. Scatter Plot with Labels and Titles

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['area'], df['price'], color='g', marker='s', s=75, label='Price vs Area')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Scatter Plot of Price vs Area with Labels')
plt.legend()
plt.show()
```

## 9. Vertical Bar Chart

```python
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['price'], color='b', edgecolor='black', label='Price')
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Vertical Bar Chart of Prices')
plt.legend()
plt.show()
```

## 10. Horizontal Bar Chart

```python
plt.figure(figsize=(10, 6))
plt.barh(df.index, df['price'], color='r', edgecolor='black', label='Price')
plt.xlabel('Price')
plt.ylabel('Index')
plt.title('Horizontal Bar Chart of Prices')
plt.legend()
plt.show()
```

## 11. Customized Bar Chart

```python
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['price'], color='g', edgecolor='black', label='Price')
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Customized Bar Chart of Prices')
plt.legend()
plt.show()
```

## 12. Basic Pie Chart

```python
plt.figure(figsize=(10, 6))
plt.pie(df['bedrooms'].value_counts(), labels=df['bedrooms'].value_counts().index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Bedroom Counts')
plt.show()
```

## 13. Customized Pie Chart

```python
plt.figure(figsize=(10, 6))
plt.pie(df['bedrooms'].value_counts(), labels=df['bedrooms'].value_counts().index, autopct='%1.1f%%', startangle=140, colors=['red', 'green', 'blue', 'yellow'])
plt.title('Customized Pie Chart of Bedroom Counts')
plt.legend(title='Bedrooms')
plt.show()
```

## 14. Exploded Pie Chart

```python
bedroom_counts = df['bedrooms'].value_counts()
explode = [0] * len(bedroom_counts)
max_index = bedroom_counts.idxmax()
explode[bedroom_counts.index.get_loc(max_index)] = 0.1

plt.figure(figsize=(10, 6))
plt.pie(bedroom_counts, labels=bedroom_counts.index, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Exploded Pie Chart of Bedroom Counts')
plt.show()
```

## 15. Pie Chart with Percentage Labels

```python
plt.figure(figsize=(10, 6))
plt.pie(df['bedrooms'].value_counts(), labels=df['bedrooms'].value_counts().index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Bedroom Counts with Percentage Labels')
plt.show()
```
